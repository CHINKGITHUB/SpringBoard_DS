```python
# 2. drop irrelevant columns, leaving only "object_id", "opted_in_to_mailing_list", "enabled_for_marketing_drip" and 'adopted_user'

users.drop(['creation_time', 'name', 'email', 'creation_source', 'last_session_creation_time', 'org_id', 'invited_by_user_id'], axis=1, inplace=True)
users
```

```python
# 3. Split test and Train data

train = users[:9600] #Training Set
test = users[9600:] #Test Set (based off of Test set being 20% of whole dataset)

df_dummies = pd.get_dummies(users)
X = df_dummies.drop(columns = ['adopted_user'])
```

```python
#Drop Adopted user columns from Train and test

trainx = train.drop('adopted_user', axis = 1)

trainy = train['adopted_user']
```

```python
#Setting up Testing Dataset

testx = test.drop('adopted_user', axis = 1)

testy = test['adopted_user']
```

```python
# 4. Feed these into .fit to build models

##.fit to build Logistic Regression (L.R.) Model

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.metrics import roc_auc_score,roc_curve,scorer

logit = LogisticRegression(class_weight = 'balanced').fit(trainx, trainy)

logit.fit(trainx,trainy)
predict = logit.predict(testx)
Proba  = logit.predict_proba(testx)

print("Algorithm :",  logit)
print ("\n Classification report : \n",classification_report(testy,predict))
print ("Accuracy   Score : ",accuracy_score(testy,predict))
model_roc_auc = roc_auc_score(testy,predict) 
print ("Area under curve : ",model_roc_auc,"\n")
fpr,tpr,thresholds = roc_curve(testy,Proba[:,1])
```

Conclusions from Summary Tables:
Precision identifies the frequency with which a model was correct when predicting the positive class Recall answers the following question: Out of all the possible positive labels, how many did the model correctly identify? F1-score is the weighted average of the precision and recall, where an F1 score reaches its best value at 1 and worst score at 0.
For Classification model metrics/assessment, values closest to 1 were most desirable signaling correct positive predictions. Model that had mostly 1's was Decision Tree model.
Therefore, to re-assess this data, I would choose the Decision Tree model due to it having a high success rate in predicting which users would be classified as 'adopted user'.
